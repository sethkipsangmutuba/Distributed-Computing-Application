## 2.1.1 Layered Architectures

### **Definition**
A design where system components are arranged in layers, each offering services to the layer above and relying on the layer below.

### **Key Properties**
- Clear separation of concerns  
- Supports portability  
- Simplifies maintenance  
- Enhances scalability

- **Application Layer** – end-user programs  
- **Middleware Layer** – communication, naming, security  
- **Operating System Layer** – resource management  
- **Hardware Layer** – physical devices  

### **Advantages**
- Easy to update/replace a layer  
- High modularity  

### **Disadvantages**
- Potential performance overhead  
- Rigid layer boundaries  

---

### **Definition**
An architecture where system functionality is offered as independent services that communicate over standardized interfaces.

### **Characteristics**
- Services are loosely coupled  
- Uses standard protocols (HTTP, SOAP, REST)  
- Focus on reusability and interoperability  

### **Key Components**
- Service provider  
- Service consumer  
- Service registry (optional)  

### **Advantages**
- Easy integration  
- High flexibility  
- Supports multiple platforms  

### **Disadvantages**
- Network overhead  
- Requires strict interface definitions  

---

## 2.1.3 Publish–Subscribe Architectures

### **Definition**
A communication style where publishers send messages to topics/channels, and subscribers receive messages for topics they’re interested in.

### **Characteristics**
- Asynchronous communication  
- Fully decoupled components  
- Scalable for large distributed systems  

### **Advantages**
- Easy to add/remove nodes  
- High scalability  

### **Disadvantages**
- Harder to guarantee message ordering  
- Debugging is more complex

## **2.1.1 Layered Architectures**

A layered architecture organizes components into tiers where higher layers call lower ones.  
Widely used in telecom systems due to clarity, modularity, and easier upgrades.

---

### **Layer Interaction**

Three patterns exist:

#### **1. Pure Layering**
Each layer only calls the next lower one.  
Common in protocol stacks (TCP/IP, OSI).

#### **2. Mixed Layering**
Higher layers may call multiple lower layers.  
Used in telecom apps accessing OS + crypto + protocol libraries.

#### **3. Upcalls**
Lower layers notify higher ones.  
Examples:
- OS interrupts  
- Network event callbacks  

---

### **Layered Communication Protocols (Telecom Context)**

Telecommunications heavily depends on layered protocol stacks:

- **OSI model**
- **TCP/IP**
- **5G SBA (Service-Based Architecture)**

Each layer provides:
- **A service** (e.g., reliable transmission)
- **An interface** (API exposed to layers above)
- **A protocol** (implementation of the service)

**Example**
- *Service*: Reliable, ordered transmission  
- *Protocol*: TCP  
- *Interface*: `socket(), connect(), send(), recv()`

This separation is fundamental in telecom networks where physical, link, network, transport, and application layers must remain modular.

---

This layered architectural model is widely used in telecom services such as:

# EXAMPLE: SERVER + CLIENT

````python
# --------------------------------------------
# EXAMPLE: SERVER + CLIENT
# --------------------------------------------
# NOTE:
# This script runs the server first, then the client.
# In real deployments, run server and client in different terminals.
# --------------------------------------------

from socket import *
import threading
import time

HOST = "127.0.0.1"
PORT = 5000

# ---------------------------
# SERVER LOGIC (THREAD)
# ---------------------------
def run_server():
    s = socket(AF_INET, SOCK_STREAM)
    s.bind((HOST, PORT))
    s.listen(1)
    print("[SERVER] Listening...")

    (conn, addr) = s.accept()
    print(f"[SERVER] Connected by {addr}")

    while True:
        data = conn.recv(1024)
        if not data:
            break
        msg = data.decode() + "*"  # simple processing
        conn.send(msg.encode())

    conn.close()
    s.close()
    print("[SERVER] Connection closed.")

# Start server thread
server_thread = threading.Thread(target=run_server, daemon=True)
server_thread.start()

# Give server time to start
time.sleep(1)

# ---------------------------
# CLIENT LOGIC
# ---------------------------
s = socket(AF_INET, SOCK_STREAM)
s.connect((HOST, PORT))

msg = "Hello World"
print(f"[CLIENT] Sending: {msg}")
s.send(msg.encode())

data = s.recv(1024).decode()
print(f"[CLIENT] Received: {data}")

s.close()
print("[CLIENT] Connection closed.")
````

## 2.1.2 Service-Oriented Architectures (SOA)

In distributed systems for telecom and software engineering, layered architectures can create tight dependencies between components, limiting flexibility and maintainability. SOA emerged to address this by organizing systems as **loosely coupled, independent services**.

---

### **Object-Based Architectures**

- Each component (object) encapsulates **state** and **operations**.  
- Objects communicate via **procedure calls**, potentially across networks.  
- Remote objects expose **interfaces**, while implementations stay hidden on the server.  
- **Client-side proxies** and **server stubs (skeletons)** transparently handle network communication.

**Telecom relevance:**  
Used in signaling servers, call-processing modules, network controllers—providing modularity and maintainability.

---

### **Microservices**

Microservices are small, independent services running as separate networked processes.

**Key principles:**
- Modularity  
- Independence  
- Replaceability  

**Deployment flexibility:**  
Services can run across **cloud, edge, or fog** infrastructure—important for telecom systems such as 5G core functions and MEC (Multi-access Edge Computing).

---

### **Service Composition**

SOA systems are created by composing independent services, possibly across different administrative domains.

**Example (Telecom Web Service):**
1. Local **order-processing** service  
2. External **payment** service  
3. Local **delivery-notification** service  

This enables services to operate independently yet collaboratively, supporting **scalability** and **resilience**.

---

### **Resource-Based / RESTful Architectures**

REST models distributed systems as collections of **resources**, each identified by a **URI**.

**Core operations:**
- `GET` – retrieve  
- `POST` – create  
- `PUT` – modify  
- `DELETE` – remove  

REST is:
- Stateless  
- Simple  
- Scalable  

**Telecom applications:**
- Network management APIs  
- Subscriber information retrieval  
- Cloud storage (e.g., Amazon S3)  
- Web-scale telecom platforms

---

### **Telecom/Software Engineering Implications**

- **SOA/microservices** enable flexible, modular distributed telecom systems with independent deployment cycles.
- **RESTful services** provide simple, scalable integration between distributed network functions.
- Proper **service composition** ensures cross-domain interoperability, maintainability, and system resilience.

# 2.1.3 Publish-Subscribe Architectures

Publish-subscribe (pub-sub) architectures are designed for loosely coupled distributed systems, where processes can join, leave, or fail without breaking the system. This makes them ideal for telecom networks, IoT systems, and event-driven software platforms.

---

## **Key Concepts**

### **Autonomous Processes**
Each process operates independently.

### **Coordination**
Mediates communication and cooperation among processes without tight coupling.

### **Coupling Types**
- **Direct coordination:**  
  Processes know each other and must be active simultaneously (like a phone call).

- **Mailbox coordination:**  
  Temporally decoupled; messages are stored for later retrieval.

- **Event-based coordination:**  
  Temporally coupled but referentially decoupled; processes subscribe to events.

- **Shared data spaces:**  
  Temporally and referentially decoupled; processes communicate via tuples in a shared space (e.g., Linda tuple spaces).

---

## **Event-Based and Shared Data Spaces**

### **Event-Based Systems**
Processes publish notifications; subscribers receive events they are interested in.

### **Shared Data Spaces**
Use a tuple-space model:
- **out(t):** Add a tuple (data item)  
- **rd(t):** Read a tuple (non-destructive)  
- **in(t):** Remove a tuple (destructive)  

Subscribers do not need to know the publishers, enhancing modularity and scalability.

**Example:**  
A microblog where users post messages to a shared space, and others read them without knowing the authors.

---

## **Subscription Models**

### **Topic-based**
Subscribers specify event types they are interested in (simple).

### **Content-based**
Subscribers specify conditions over attributes or ranges (more expressive, more complex).

Middleware matches subscriptions to events and forwards notifications or data.

---

## **Telecom & Software Engineering Implications**

- Enables **scalable, event-driven systems**  
  (e.g., network monitoring, alert systems, IoT sensor networks)

- **Decoupling** reduces dependencies, allowing processes to scale independently.

- Challenges include:  
  - Efficient subscription matching  
  - Event composition  
  - Handling temporal/state dependencies in large-scale networks  

Pub-sub is critical for real-time telecom applications, such as:
- Distributed signaling  
- Subscriber notifications  
- Cloud-based service orchestration  

---


````python
"""
Simulation: Telecom Publish–Subscribe Architecture
(Brief Python implementation, telecom/IoT context)
"""

# ------------------- EVENT-BASED PUB-SUB (TOPIC-BASED) -------------------

class EventBus:
    def __init__(self):
        self.subscribers = {}  # topic -> list of callbacks

    def subscribe(self, topic, callback):
        self.subscribers.setdefault(topic, []).append(callback)

    def publish(self, topic, data):
        for cb in self.subscribers.get(topic, []):
            cb(data)


# Example telecom event handlers
def on_cell_down(event):
    print(f"[OPS] Cell outage alert received: {event}")

def on_traffic_spike(event):
    print(f"[NOC] Traffic anomaly detected: {event}")


# ------------------- SHARED DATA SPACE (TUPLE SPACE) -------------------

class TupleSpace:
    def __init__(self):
        self.space = []

    def out(self, t):
        self.space.append(t)

    def rd(self, cond):
        for t in self.space:
            if cond(t):
                return t

    def inp(self, cond):
        for i, t in enumerate(self.space):
            if cond(t):
                return self.space.pop(i)


# Example: IoT sensor readings stored in tuple space
def sensor_post(space, sensor_id, value):
    space.out(("sensor", sensor_id, value))

def controller_read(space):
    return space.rd(lambda t: t[0] == "sensor")


# ------------------- SIMULATION -------------------

if __name__ == "__main__":
    # Event-based pub-sub
    bus = EventBus()
    bus.subscribe("cell_down", on_cell_down)
    bus.subscribe("traffic_spike", on_traffic_spike)

    bus.publish("cell_down", {"id": "LTE-101", "severity": "HIGH"})
    bus.publish("traffic_spike", {"sector": 3, "uplink_mbps": 95})

    # Tuple-space shared data example
    ts = TupleSpace()
    sensor_post(ts, "temp01", 32.5)
    sensor_post(ts, "hum02", 60)

    print("Controller read:", controller_read(ts))
````

# 2.2 Middleware and Distributed Systems

Middleware is a software layer positioned between distributed applications and the underlying operating systems. It provides standardized services and abstractions that simplify communication, coordination, and resource sharing across multiple networked machines.

---

## **Key Functions**

### **1. Communication**

It abstracts low-level networking details and provides uniform communication APIs.

Hides heterogeneity across:
- Hardware  
- Operating systems  
- Network protocols  

Developers interact with a consistent programming model instead of dealing with system-level complexities.

- Manages shared network resources
- Optimizes distributed resource allocation
- Helps coordinate execution across nodes

- Offering shared containers for reusable components

# 2.2.1 Middleware Organization

Middleware in distributed systems especially telecom and large-scale software platforms must be adaptable, extensible, and capable of integrating heterogeneous components. Two major design patterns support this flexibility: **wrappers** and **interceptors**.

---

## **Wrappers**

### **Purpose**
Wrappers adapt legacy or incompatible component interfaces so they can be used by modern applications.

### **Functionality**
- Transform client requests into operations supported by the underlying component.
- Hide differences between systems, protocols, or interfaces.

### **Example**
Amazon S3 provides a **RESTful wrapper** over its internal distributed storage engine.  
Applications interact using HTTP methods, even though the internal implementation is very different.

### **Scalability Challenge**
Without middleware:
- \( N \) applications and \( N \) services require **O(N²)** pairwise wrappers.

### **Solution: Message Broker**
A broker centralizes communication:
- Applications send requests to the broker.
- Broker routes, aggregates, or transforms messages.
- Reduces wrappers to **O(N)**.

**Benefits:**
- Simplified integration  
- Centralized control  
- Easy monitoring and transformation of messages  

---

## **Interceptors**

### **Purpose**
Add new behavior to middleware transparently—without modifying the core logic of applications or components.

### **Functionality**
Interceptors hook into:
- Method calls  
- Message sends/receives  
- Protocol operations  

They can perform:
- Logging  
- Replication  
- Encryption/decryption  
- Load balancing  
- Message fragmentation/reassembly for large payloads  

### **Example in Object-Based Middleware**
1. Object A calls a method on Object B using a local interface.  
2. Middleware intercepts the call and converts it into a **generic object invocation**.  
3. Call is packaged into a network message.  
4. Interceptors may add replication or fragmentation logic.  
5. Object B receives the invocation without knowing any middleware modifications.

**Benefit:**  
Applications remain unaware of replication policies, network routing, or reliability mechanisms.

---

## **Telecom & Software Engineering Implications**

### **Wrappers**
- Allow telecom operators to integrate legacy systems (e.g., SS7, Diameter, old billing platforms).
- Enable interoperability between heterogeneous components.

### **Interceptors**
- Useful for telecom-specific concerns such as:
  - QoS (Quality of Service) control  
  - Replication of signaling state  
  - Fragmentation of large media or control messages  
  - Runtime policy enforcement

Middleware organization provides:

These properties are essential for distributed telecom systems such as:

````python
"""
Telecom Middleware (Wrappers, Broker, Interceptors)
Brief Python implementation.
"""

# ------------------------ WRAPPER EXAMPLE ------------------------
# Goal: Adapt different telecom services to expose a unified interface.

class LegacySMSService:
    def send_text(self, number, txt):
        return f"[LegacySMS] Sent to {number}: {txt}"

class LegacyEmailService:
    def deliver(self, address, body):
        return f"[EmailSys] Delivered to {address}: {body}"

# Wrapper normalizes heterogeneous interfaces
class MessageWrapper:
    def __init__(self, service):
        self.service = service

    def send(self, recipient, message):
        if isinstance(self.service, LegacySMSService):
            return self.service.send_text(recipient, message)
        if isinstance(self.service, LegacyEmailService):
            return self.service.deliver(recipient, message)
        return "[Wrapper] Unsupported service"


# ------------------------ BROKER EXAMPLE ------------------------
# Central broker reduces N^2 connections to N.

class Broker:
    def __init__(self):
        self.services = {}  # name -> wrapper

    def register(self, name, wrapper):
        self.services[name] = wrapper

    def request(self, service_name, recipient, message):
        svc = self.services.get(service_name)
        if not svc:
            return "[Broker] Service not found"
        return svc.send(recipient, message)


# ------------------------ INTERCEPTORS ------------------------
# For telecom, interceptors handle replication, logging, QoS tagging, etc.

def logging_interceptor(func):
    def wrapper(*args, **kw):
        print(f"[LOG] Calling {func.__name__} with {args[1:]}")
        return func(*args, **kw)
    return wrapper

def qos_interceptor(level):
    def decorator(func):
        def wrapper(*args, **kw):
            print(f"[QoS-{level}] Tagging message before dispatch")
            return func(*args, **kw)
        return wrapper
    return decorator


# Apply interceptors to a broker method
class InterceptedBroker(Broker):
    @logging_interceptor
    @qos_interceptor(level="HIGH")
    def request(self, service_name, recipient, message):
        return super().request(service_name, recipient, message)


# ------------------------ SIMULATION ------------------------

if __name__ == "__main__":
    sms = MessageWrapper(LegacySMSService())
    email = MessageWrapper(LegacyEmailService())

    broker = InterceptedBroker()
    broker.register("sms", sms)
    broker.register("email", email)

    print(broker.request("sms", "+254700112233", "Cell down in sector 3"))
    print(broker.request("email", "noc@telco.com", "Traffic spike detected"))
````

# 2.2.2 Modifiable Middleware

Middleware in distributed systems must adapt to changing environments, especially in telecom and large-scale software platforms. These environments involve fluctuating network conditions, device mobility, hardware failures, and varying resource constraints such as battery life. Modifiable middleware addresses these concerns by supporting dynamic adjustments without requiring application-level changes.

---

## **Key Concepts**

### **Adaptive Middleware**
Adaptive middleware can modify its behavior at runtime in response to environmental or system-level changes.

- Automatically reacts to variations in:
  - Network bandwidth
  - Latency
  - Device mobility
  - System failures
- Uses mechanisms such as **interceptors** and **wrappers** to adjust message flow or interface behavior dynamically.
- Allows applications to remain simple and focused on business logic.

---

## **Component-Based Design**

Modern middleware is built from **modular components** that can be:

- Loaded
- Unloaded
- Replaced  
all *while the system is running*.

### **Requirements**
Each component must clearly specify:

- **Provided interfaces**
- **Required interfaces**
- **How state is managed** between calls

This modular design enables:

- Dynamic reconfiguration
- Late binding
- Targeted upgrades without full system restarts

---

## **Modifiability Challenges**

Updating one component may cause cascading effects across other components.

Challenges include:

- Managing complex inter-component dependencies  
- Ensuring consistent state across distributed nodes  
- Updating components in multi-node systems without service disruption  

These challenges are significantly harder in distributed systems than in standalone systems.

---

## **Implications for Telecommunications & Software Engineering**

### **Telecommunication Networks**
Modifiable middleware is essential for:

- Handling variable bandwidth and latency  
- Managing mobility in wireless networks  
- Ensuring uninterrupted service during upgrades  

Examples:  
5G core network functions, virtualized network services (VNFs), and real-time signaling systems.

### **Cloud and Edge Computing**
Supports:

Examples:  
Kubernetes-controlled microservices, edge-based telecom functions (e.g., UPF in 5G).

### **Resilience and Maintainability**
- Enables high availability  
- Allows rapid adaptation to failures or traffic spikes  
- Reduces downtime for maintenance

---

Modifiable middleware provides the flexibility required for modern distributed systems, especially large-scale telecom environments. Its ability to adapt dynamically, update components on the fly, and hide complexity from applications makes it foundational for robust, scalable, and maintainable distributed architectures.



````python
import time
import random
import threading

# ----------------------------
# Base Middleware Component
# ----------------------------
class MiddlewareComponent:
    def __init__(self, name):
        self.name = name
        self.state = {}
        self.active = True

    def process(self, message):
        if not self.active:
            return message
        # Default behavior: just log message
        print(f"[{self.name}] Processing message: {message}")
        return message

    def update_state(self, key, value):
        self.state[key] = value

    def shutdown(self):
        self.active = False
        print(f"[{self.name}] Component deactivated.")

    def activate(self):
        self.active = True
        print(f"[{self.name}] Component activated.")

# ----------------------------
# Modifiable Middleware System
# ----------------------------
class ModifiableMiddleware:
    def __init__(self):
        self.components = {}

    def add_component(self, comp: MiddlewareComponent):
        self.components[comp.name] = comp
        print(f"[Middleware] Added component: {comp.name}")

    def remove_component(self, name):
        if name in self.components:
            self.components[name].shutdown()
            del self.components[name]
            print(f"[Middleware] Removed component: {name}")

    def replace_component(self, name, new_comp):
        self.remove_component(name)
        self.add_component(new_comp)
        print(f"[Middleware] Replaced component: {name} with {new_comp.name}")

    def process_message(self, message):
        for comp in self.components.values():
            message = comp.process(message)
        return message

# ----------------------------
# Adaptive Network Component
# ----------------------------
class AdaptiveNetwork(MiddlewareComponent):
    def process(self, message):
        # Simulate network fluctuations
        latency = random.uniform(0.01, 0.5)  # seconds
        bandwidth = random.choice([10, 50, 100])  # Mbps
        print(f"[{self.name}] Network conditions -> latency: {latency:.3f}s, bandwidth: {bandwidth}Mbps")
        time.sleep(latency)
        return super().process(f"{message} | Latency:{latency:.3f}s, BW:{bandwidth}Mbps")

# ----------------------------
# Simulate Telecom Messages
# ----------------------------
def telecom_simulation(middleware: ModifiableMiddleware, num_messages=5):
    for i in range(num_messages):
        message = f"Packet-{i+1}"
        print(f"\n[Simulation] Sending: {message}")
        middleware.process_message(message)
        time.sleep(0.2)

# ----------------------------
# Runtime Dynamic Update Example
# ----------------------------
middleware = ModifiableMiddleware()
comp1 = AdaptiveNetwork("NetworkMonitor")
comp2 = MiddlewareComponent("LoggingComponent")

middleware.add_component(comp1)
middleware.add_component(comp2)

# Start simulation in a separate thread to allow dynamic updates
simulation_thread = threading.Thread(target=telecom_simulation, args=(middleware,))
simulation_thread.start()

# Dynamic modification at runtime
time.sleep(0.7)
new_comp = MiddlewareComponent("SecurityComponent")
middleware.replace_component("LoggingComponent", new_comp)

simulation_thread.join()
````

# 2.3.1 Simple Client-Server Architecture

## Core Idea
A distributed system divides processes into **clients** and **servers**:

- **Server:** Provides a specific service (e.g., file system, database).  
- **Client:** Requests services from a server and waits for a reply.  

Interaction follows a **request-reply pattern**.

---

## Communication Protocols

### Connectionless Protocols (UDP-like)
- Efficient in reliable networks (e.g., LANs).  
- Client sends a request; server replies.  
- **Issue:** Message loss → client may resend request, risking duplicate operations.  
  - **Idempotent operations:** Safe to repeat (e.g., “check balance”)  
  - **Non-idempotent operations:** Risky to repeat (e.g., “transfer money”)

### Connection-Oriented Protocols (TCP-like)
- Ensures reliable communication over unreliable networks (e.g., WANs, Internet).  
- Flow: Client establishes connection → sends request → server replies → connection closed.  
- **Drawback:** Overhead of connection setup and teardown, especially for small requests.

---

## Practical Notes
- **Client-server roles can overlap:**  
  A database server may act as a client to underlying file servers while serving queries to other clients.  
- **Idempotency awareness** is critical in telecom and financial software to prevent errors during network failures.

---

## Relevance to Telecom & Software Engineering
- **Telecom networks:** Signaling servers act as clients to database servers and as servers to mobile devices.  
- **Software systems:** Web applications, microservices, and cloud APIs follow this model, often using TCP/IP for reliability.  

**Design Implication:** Must carefully choose protocol type based on network reliability, operation type, and system performance requirements.


````python
import socket
import threading
import time
import random

# ----------------------------
# Server Implementations
# ----------------------------

# TCP-like server (connection-oriented)
def tcp_server(host='127.0.0.1', port=5000):
    server_sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server_sock.bind((host, port))
    server_sock.listen(5)
    print(f"[TCP Server] Listening on {host}:{port}")

    while True:
        conn, addr = server_sock.accept()
        data = conn.recv(1024).decode()
        print(f"[TCP Server] Received from {addr}: {data}")
        # Simulate non-idempotent operation
        if "transfer" in data:
            response = f"Processed {data} once."
        else:
            response = f"Acknowledged: {data}"
        conn.send(response.encode())
        conn.close()

# UDP-like server (connectionless)
def udp_server(host='127.0.0.1', port=5001):
    server_sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
    server_sock.bind((host, port))
    print(f"[UDP Server] Listening on {host}:{port}")

    while True:
        data, addr = server_sock.recvfrom(1024)
        data = data.decode()
        print(f"[UDP Server] Received from {addr}: {data}")
        # Simulate message loss (10%)
        if random.random() < 0.1:
            print(f"[UDP Server] Dropped message from {addr}")
            continue
        response = f"Acknowledged: {data}"
        server_sock.sendto(response.encode(), addr)

# ----------------------------
# Client Implementations
# ----------------------------

# TCP-like client
def tcp_client(message, host='127.0.0.1', port=5000):
    client_sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    client_sock.connect((host, port))
    client_sock.send(message.encode())
    response = client_sock.recv(1024).decode()
    print(f"[TCP Client] Received: {response}")
    client_sock.close()

# UDP-like client
def udp_client(message, host='127.0.0.1', port=5001, retries=3):
    client_sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
    for attempt in range(retries):
        client_sock.sendto(message.encode(), (host, port))
        client_sock.settimeout(0.5)
        try:
            response, _ = client_sock.recvfrom(1024)
            print(f"[UDP Client] Received: {response.decode()}")
            break
        except socket.timeout:
            print(f"[UDP Client] Timeout, retry {attempt+1}")
    client_sock.close()

# ----------------------------
# Simulation of Telecom-like Requests
# ----------------------------

# Start servers in separate threads
threading.Thread(target=tcp_server, daemon=True).start()
threading.Thread(target=udp_server, daemon=True).start()
time.sleep(1)  # Allow servers to start

# Simulate TCP clients (connection-oriented)
tcp_client("check_balance")
tcp_client("transfer_money")  # Non-idempotent

# Simulate UDP clients (connectionless)
udp_client("status_update")
udp_client("location_ping")
````

# 2.3.2 Multitiered Architectures

## Concept
Extends the client-server model by logically dividing applications into three layers:

- **User-interface layer:** Handles presentation.  
- **Processing layer:** Implements business logic.  
- **Data layer:** Manages storage and retrieval.  

These layers can be physically distributed across machines, resulting in **two-tier** or **three-tier architectures**.

---

## Two-Tier Architecture
- **Client machine:** Runs part of the user interface.  
- **Server machine:** Handles the rest (processing + data).  

**Variants:**
- **Thin client:** Only terminal-dependent UI on client; server handles all processing.  
- **Full client UI:** Entire UI on client; communicates with server via protocol.  
- **Partial application on client:** Client performs pre-processing or checks (e.g., form validation, basic text editing).  
- **Fat client with server storage:** Most processing on client; server stores data.  
- **Client-side caching:** Client maintains local copies of frequently accessed data.  

**Trade-offs:**
- **Thin clients:** Easier to manage, less platform-dependent, lower maintenance.  
- **Fat clients:** Better offline operation, local performance, richer UI, but harder to manage.

---

## Three-Tier Architecture
- Introduces **application servers** between client and database servers.  
- **Architecture flow:**  
  1. Client requests → application server  
  2. Application server processes logic → database server  
  3. Database server returns results → application server → client  

**Examples:**
- Transaction processing: Transaction monitor coordinates multiple data servers.  
- Web applications: Web server (entry point) → application server (processing) → database server (data storage).  

**Advantages:**
- Better **scalability** (server-side tasks distributed across multiple machines).  
- Easier **maintenance and updates**, since logic is centralized.  
- Clients can remain **thin**, reducing platform dependency and management overhead.

**Modern trends:**
- Even thin-client architectures rely on distributed server-side solutions, e.g., cloud computing.  
- Virtual desktops and advanced web technologies allow sophisticated client-side functionality without fat-client drawbacks.


````python
import threading
import time
import random
from queue import Queue

# ----------------------------
# Simulated Data Layer (Database Server)
# ----------------------------
class DatabaseServer:
    def __init__(self):
        self.data_store = {"user1": 100, "user2": 250}

    def query(self, user):
        print(f"[DB Server] Retrieving balance for {user}")
        time.sleep(random.uniform(0.1, 0.3))  # Simulate DB latency
        return self.data_store.get(user, "User not found")

    def update(self, user, amount):
        print(f"[DB Server] Updating balance for {user}")
        self.data_store[user] = amount
        return f"Updated {user}'s balance to {amount}"

# ----------------------------
# Application Server (Middle Tier)
# ----------------------------
class ApplicationServer:
    def __init__(self, db_server):
        self.db = db_server

    def process_request(self, request):
        print(f"[App Server] Processing request: {request}")
        user, action, *args = request
        if action == "query":
            return self.db.query(user)
        elif action == "update":
            return self.db.update(user, args[0])
        else:
            return "Invalid request"

# ----------------------------
# Client Layer
# ----------------------------
class Client:
    def __init__(self, app_server=None, db_server=None, mode="thin"):
        self.app_server = app_server
        self.db_server = db_server
        self.mode = mode  # "thin" or "fat"

    def send_request(self, request):
        if self.mode == "thin" and self.app_server:
            # Thin client: sends to app server
            response = self.app_server.process_request(request)
            print(f"[Thin Client] Response: {response}")
        elif self.mode == "fat" and self.db_server:
            # Fat client: processes locally or queries DB directly
            action = request[1]
            if action == "query":
                response = self.db_server.query(request[0])
            elif action == "update":
                response = self.db_server.update(request[0], request[2])
            else:
                response = "Invalid request"
            print(f"[Fat Client] Response: {response}")
        else:
            print("[Client] No server available for request")

# ----------------------------
# Two-Tier Architecture Simulation
# ----------------------------
print("\n=== Two-Tier Architecture Simulation ===")
db_server = DatabaseServer()
thin_client = Client(db_server=db_server, mode="fat")  # Fat client directly queries DB

# Fat client example
thin_client.send_request(("user1", "query"))
thin_client.send_request(("user2", "update", 300))

# ----------------------------
# Three-Tier Architecture Simulation
# ----------------------------
print("\n=== Three-Tier Architecture Simulation ===")
app_server = ApplicationServer(db_server)
thin_client = Client(app_server=app_server, mode="thin")  # Thin client uses app server

# Thin client example
thin_client.send_request(("user1", "query"))
thin_client.send_request(("user2", "update", 400))
````

# 2.3.3 Ex1: Network File System (NFS)

## Concept
- NFS is a **distributed file system** following a client-server architecture, widely used in Unix systems.  
- **Goal:** Provide a standardized view of local file systems so clients can access files transparently, regardless of the underlying file system implementation.

---

## Key Models of File Access

### Remote Access Model (Figure 2.19a)
- Files stay on the **server**.  
- Client requests access; operations are executed remotely on the server.  
- Transparent to the client: interface resembles a **local file system**.

### Upload/Download Model (Figure 2.19b)
- Files are downloaded to the **client** for local access.  
- After modification, files are uploaded back to the server.  
- Example: FTP usage.

---

## NFS Architecture for Unix Systems (Figure 2.20)

### Client-side
- Local applications use **system calls**.  
- Calls are redirected to the **Virtual File System (VFS)** layer → a standard interface for various file systems.  
- **NFS client** converts file operations into **Remote Procedure Calls (RPCs)** to communicate with the server.

### Server-side
- **NFS server** receives RPC requests.  
- RPC server stub converts requests to VFS operations.  
- VFS interacts with the local file system where the actual files reside.

**Advantage:** NFS is independent of the underlying local file system; the server and client OS can differ as long as they support NFS conventions.  
**Limitation:** Local file system constraints (e.g., MS-DOS short filenames) may limit full transparency.

---

## Key Points
- NFS allows **heterogeneous processes** across different OSs to share a common file system.  
- Uses **RPC** as the communication mechanism between clients and servers.  
- The **VFS layer** abstracts differences in file system implementations.


````python
import threading
import time
import random
from queue import Queue

# ----------------------------
# Simulated NFS Server
# ----------------------------
class NFSServer:
    def __init__(self):
        # Simulated server-side file system
        self.files = {
            "file1.txt": "Hello from NFS Server",
            "file2.txt": "Distributed Systems Rocks!"
        }
        self.lock = threading.Lock()

    def read_file(self, filename):
        with self.lock:
            content = self.files.get(filename, "File not found")
            print(f"[NFS Server] Read '{filename}': {content}")
            time.sleep(random.uniform(0.1, 0.3))  # Simulate network latency
            return content

    def write_file(self, filename, content):
        with self.lock:
            self.files[filename] = content
            print(f"[NFS Server] Written '{filename}': {content}")
            time.sleep(random.uniform(0.1, 0.3))
            return f"Write successful: {filename}"

# ----------------------------
# Simulated NFS Client
# ----------------------------
class NFSClient:
    def __init__(self, server: NFSServer, mode="remote"):
        self.server = server
        self.mode = mode  # "remote" or "download"
        self.local_cache = {}

    def read(self, filename):
        if self.mode == "remote":
            # Remote access model
            return self.server.read_file(filename)
        elif self.mode == "download":
            # Upload/Download model
            content = self.server.read_file(filename)
            self.local_cache[filename] = content
            print(f"[NFS Client] Cached '{filename}' locally")
            return content

    def write(self, filename, content):
        if self.mode == "remote":
            return self.server.write_file(filename, content)
        elif self.mode == "download":
            # Update local cache and upload back
            self.local_cache[filename] = content
            print(f"[NFS Client] Updated local cache '{filename}'")
            return self.server.write_file(filename, content)

# ----------------------------
# Simulation Example
# ----------------------------
nfs_server = NFSServer()

# Remote access client
client_remote = NFSClient(nfs_server, mode="remote")
print("\n--- Remote Access Model ---")
print(client_remote.read("file1.txt"))
client_remote.write("file1.txt", "Updated remotely via NFS Client")

# Upload/Download client
client_download = NFSClient(nfs_server, mode="download")
print("\n--- Upload/Download Model ---")
print(client_download.read("file2.txt"))
client_download.write("file2.txt", "Modified locally and uploaded back")
````

````python
import socket
import threading
import time
import random
from queue import Queue
from flask import Flask, request, jsonify
import requests

# ----------------------------
# TCP Server (Connection-Oriented)
# ----------------------------
def tcp_server(host='127.0.0.1', port=5000):
    server_sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server_sock.bind((host, port))
    server_sock.listen(5)
    print(f"[TCP Server] Listening on {host}:{port}")
    while True:
        conn, addr = server_sock.accept()
        data = conn.recv(1024).decode()
        print(f"[TCP Server] Received from {addr}: {data}")
        # Simulate non-idempotent operation
        response = f"Processed {data}"
        conn.send(response.encode())
        conn.close()

# ----------------------------
# UDP Server (Connectionless)
# ----------------------------
def udp_server(host='127.0.0.1', port=5001):
    server_sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
    server_sock.bind((host, port))
    print(f"[UDP Server] Listening on {host}:{port}")
    while True:
        data, addr = server_sock.recvfrom(1024)
        data = data.decode()
        if random.random() < 0.1:  # Simulate message loss
            print(f"[UDP Server] Dropped message from {addr}")
            continue
        print(f"[UDP Server] Received from {addr}: {data}")
        server_sock.sendto(f"Acknowledged: {data}".encode(), addr)

# ----------------------------
# Database and Application Server (Three-Tier)
# ----------------------------
class DatabaseServer:
    def __init__(self):
        self.data_store = {"user1": 100, "user2": 250}

    def query(self, user):
        time.sleep(random.uniform(0.1,0.3))
        return self.data_store.get(user, "User not found")

    def update(self, user, amount):
        self.data_store[user] = amount
        return f"Updated {user}'s balance to {amount}"

class ApplicationServer:
    def __init__(self, db_server):
        self.db = db_server

    def process_request(self, request):
        user, action, *args = request
        if action == "query":
            return self.db.query(user)
        elif action == "update":
            return self.db.update(user, args[0])
        return "Invalid request"

class Client:
    def __init__(self, app_server=None, db_server=None, mode="thin"):
        self.app_server = app_server
        self.db_server = db_server
        self.mode = mode

    def send_request(self, request):
        if self.mode=="thin" and self.app_server:
            response = self.app_server.process_request(request)
        elif self.mode=="fat" and self.db_server:
            action = request[1]
            if action=="query":
                response = self.db_server.query(request[0])
            elif action=="update":
                response = self.db_server.update(request[0], request[2])
            else:
                response = "Invalid request"
        else:
            response = "No server available"
        print(f"[Client-{self.mode}] Response: {response}")

# ----------------------------
# NFS Simulation
# ----------------------------
class NFSServer:
    def __init__(self):
        self.files = {"file1.txt":"Hello NFS", "file2.txt":"Distributed Systems"}
        self.lock = threading.Lock()

    def read_file(self, filename):
        with self.lock:
            time.sleep(random.uniform(0.1,0.3))
            return self.files.get(filename,"File not found")

    def write_file(self, filename, content):
        with self.lock:
            self.files[filename]=content
            return f"Write successful: {filename}"

class NFSClient:
    def __init__(self, server: NFSServer, mode="remote"):
        self.server = server
        self.mode = mode
        self.local_cache = {}

    def read(self, filename):
        if self.mode=="remote":
            return self.server.read_file(filename)
        else:
            content = self.server.read_file(filename)
            self.local_cache[filename]=content
            return content

    def write(self, filename, content):
        if self.mode=="remote":
            return self.server.write_file(filename, content)
        else:
            self.local_cache[filename]=content
            return self.server.write_file(filename, content)

# ----------------------------
# Flask Web Server
# ----------------------------
app = Flask(__name__)
web_db = {"users":{"user1":"Alice","user2":"Bob"},"messages":["Welcome!","Dynamic content ready."]}

@app.route('/static/<filename>')
def static_file(filename):
    static_files = {"index.html":"<h1>Static Page</h1>","about.html":"<p>Web distributed system.</p>"}
    return static_files.get(filename,"File not found")

@app.route('/dynamic/<username>')
def dynamic_content(username):
    user_name = web_db["users"].get(username,"Guest")
    html=f"<h1>Hello, {user_name}</h1><ul>"
    for msg in web_db["messages"]:
        html+=f"<li>{msg} | {time.strftime('%H:%M:%S')}</li>"
    html+="</ul>"
    return html

@app.route('/api/get_message', methods=['GET'])
def api_message():
    index=int(request.args.get("index",0))
    return jsonify({"message":web_db["messages"][index % len(web_db["messages"])],"timestamp":time.time()})

def run_web_server():
    app.run(port=5050, debug=False, use_reloader=False)

# ----------------------------
# Start all servers
# ----------------------------
threading.Thread(target=tcp_server, daemon=True).start()
threading.Thread(target=udp_server, daemon=True).start()
threading.Thread(target=run_web_server, daemon=True).start()
time.sleep(1)

# ----------------------------
# Simulate clients
# ----------------------------
# TCP client
def tcp_client(message, host='127.0.0.1', port=5000):
    sock=socket.socket(socket.AF_INET,socket.SOCK_STREAM)
    sock.connect((host,port))
    sock.send(message.encode())
    print(f"[TCP Client] Received: {sock.recv(1024).decode()}")
    sock.close()

tcp_client("check_balance")
tcp_client("transfer_money")

# UDP client
def udp_client(message, host='127.0.0.1', port=5001, retries=3):
    sock=socket.socket(socket.AF_INET,socket.SOCK_DGRAM)
    for i in range(retries):
        sock.sendto(message.encode(),(host,port))
        sock.settimeout(0.5)
        try:
            resp,_=sock.recvfrom(1024)
            print(f"[UDP Client] Received: {resp.decode()}")
            break
        except:
            print(f"[UDP Client] Timeout, retry {i+1}")
    sock.close()

udp_client("status_update")
udp_client("location_ping")

# Two-tier / Three-tier clients
db_server = DatabaseServer()
app_server = ApplicationServer(db_server)
thin_client = Client(app_server=app_server, mode="thin")
fat_client = Client(db_server=db_server, mode="fat")
thin_client.send_request(("user1","query"))
fat_client.send_request(("user2","update",500))

# NFS clients
nfs_server = NFSServer()
client_remote = NFSClient(nfs_server, mode="remote")
client_download = NFSClient(nfs_server, mode="download")
print("[NFS Remote] Read:", client_remote.read("file1.txt"))
client_download.write("file2.txt","Updated locally & uploaded")

# Web client requests
resp = requests.get("http://127.0.0.1:5050/static/index.html")
print("[Web Client] Static content:\n",resp.text[:50],"...")
resp = requests.get("http://127.0.0.1:5050/dynamic/user1")
print("[Web Client] Dynamic content:\n",resp.text[:50],"...")
resp = requests.get("http://127.0.0.1:5050/api/get_message",params={"index":1})
print("[Web Client] API response:\n",resp.json())
````

# 2.4 Symmetrically Distributed System Architectures

## Vertical vs Horizontal Distribution

### Vertical Distribution
- Arises naturally from **multitiered client-server architectures**.  
- Divides applications into:
  - User interface layer
  - Processing components
  - Data-management components  
- Logical components are placed on **different machines**.  
- Analogous to **vertical fragmentation** in distributed databases (splitting tables column-wise across machines).  
- **Benefit:** Logical and physical separation simplifies management and optimizes machine-specific functions.

### Horizontal Distribution
- Focuses on distributing **clients or servers** rather than separating layers.  
- Each component may be split into **logically equivalent parts**, each handling its share of the data.  
- **Goal:** Load balancing and scalability.  
- Often used in modern distributed architectures like **peer-to-peer (P2P) systems**.

---

## Peer-to-Peer (P2P) Systems
- **Symmetric behavior:** All processes are equal.  
- Each node acts as **client and server simultaneously** (also called a servant).  
- **Overlay network:** Logical network formed by processes (nodes) and communication channels (links).  
- Nodes may not communicate directly with all others; messages can traverse intermediate nodes.

### Types of Overlay Networks
- **Structured:** Nodes and data placement follow a well-defined structure (e.g., Distributed Hash Tables).  
- **Unstructured:** Nodes connect arbitrarily; data lookup often uses flooding or random walks.

### Key References and Surveys
- Lua et al., 2005 – Surveys structured and unstructured overlays.  
- Buford & Yu, 2010 – Extensive list of P2P systems.  
- Aberer et al., 2005 – Reference architecture for comparing P2P systems.  
- Androutsellis-Theotokis & Spinellis, 2004 – P2P content distribution survey.  
- Buford et al., 2009; Tarkoma, 2010; Vu et al., 2010 – Textbooks for deeper study.

---

## Key Points
- **Vertical distribution** is layer-oriented (client-server tiers).  
- **Horizontal distribution** is peer-oriented (load-balanced, symmetric).  
- **P2P systems** exemplify symmetrical architecture: equality, shared responsibilities, and decentralized control.


# 2.4.1 Structured Peer-to-Peer Systems

**Overview:**  
Structured P2P systems organize nodes (processes) in a deterministic overlay topology such as a ring, binary tree, grid, or hypercube. The main goal is to enable efficient and deterministic data lookup.

**Key Concepts:**  
- **Semantic-Free Indexing:** Each data item is assigned a unique key using a hash function:  
  `key(data item) = hash(data item value)`  
- Nodes store (key, value) pairs in a Distributed Hash Table (DHT) and are responsible for subsets of keys.  
- Lookup requests are forwarded through the overlay network to the node responsible for the requested key.  
- The efficiency of lookups relies heavily on the overlay topology.

**Examples of Overlay Topologies:**  
- **Hypercube:** Nodes are vertices in an n-dimensional cube. Lookup requests are forwarded progressively to neighbors closer to the target node.  
  *Example:* Node `0111` wants data at key `1110` → forwards to `0110` → `0110` forwards to `1110` → target responds.  

- **Chord System:** Nodes are arranged in a logical ring. Each key `k` maps to the node with the smallest identifier ≥ k (`succ(k)`). Nodes maintain shortcuts to accelerate lookups. Lookup complexity is `O(log N)` hops for N nodes.  
  *Example:* Node 9 looks for key 3 (responsible node is 4) → forwards along shortcuts → request reaches node 4.  

**Node Join/Leave in Chord:**  
- **Joining:** A new node contacts an existing node, finds its successor, inserts itself in the ring, updates shortcuts, and receives relevant data.  
- **Leaving:** The node informs its predecessor and successor, transfers its data, and updates shortcuts.

**Key Takeaways:**  
- Structured P2P systems provide deterministic and efficient lookups using DHTs.  
- Overlay topology and shortcut connections are crucial for scalability and performance.  
- Chord is a canonical example widely referenced in structured P2P research.


````python
import hashlib
import random

# ----------------------------
# Helper: Hash function for keys
# ----------------------------
def hash_key(value, m=4):
    """Return a hash key in [0, 2^m)"""
    h = int(hashlib.sha1(str(value).encode()).hexdigest(), 16)
    return h % (2 ** m)

# ----------------------------
# Node class in Chord
# ----------------------------
class ChordNode:
    def __init__(self, node_id, m=4):
        self.node_id = node_id
        self.m = m
        self.successor = self
        self.predecessor = self
        self.finger_table = [self for _ in range(m)]
        self.data = {}  # key-value storage

    def find_successor(self, key):
        if self.node_id < key <= self.successor.node_id or self.node_id >= self.successor.node_id and (key > self.node_id or key <= self.successor.node_id):
            return self.successor
        else:
            n0 = self.closest_preceding_node(key)
            if n0 == self:
                return self
            return n0.find_successor(key)

    def closest_preceding_node(self, key):
        for i in reversed(range(self.m)):
            finger_id = self.finger_table[i].node_id
            if self.node_id < finger_id < key or (self.node_id > key and self.node_id < finger_id) or (finger_id < key < self.node_id):
                return self.finger_table[i]
        return self

    def join(self, existing_node):
        if existing_node:
            self.init_finger_table(existing_node)
            self.update_others()
        else:
            for i in range(self.m):
                self.finger_table[i] = self
            self.predecessor = self
            self.successor = self

    def init_finger_table(self, existing_node):
        self.finger_table[0] = existing_node.find_successor((self.node_id + 1) % (2 ** self.m))
        self.successor = self.finger_table[0]
        self.predecessor = self.successor.predecessor
        self.successor.predecessor = self
        self.predecessor.successor = self
        for i in range(self.m - 1):
            start = (self.node_id + 2**i) % (2**self.m)
            if self.node_id < start <= self.finger_table[i].node_id or self.node_id >= self.finger_table[i].node_id and (start > self.node_id or start <= self.finger_table[i].node_id):
                self.finger_table[i+1] = self.finger_table[i]
            else:
                self.finger_table[i+1] = existing_node.find_successor(start)

    def update_others(self):
        # Simplified: real Chord updates predecessors and finger tables in other nodes
        pass

    def store(self, key, value):
        node = self.find_successor(key)
        node.data[key] = value
        print(f"[Chord] Key {key} stored at Node {node.node_id}")

    def lookup(self, key):
        node = self.find_successor(key)
        value = node.data.get(key, None)
        print(f"[Chord] Lookup key {key} found at Node {node.node_id}: {value}")
        return value

# ----------------------------
# Simulate a Chord Ring
# ----------------------------
m = 4  # 2^m keyspace
nodes = [ChordNode(i) for i in [0, 3, 7, 12]]
# Initialize ring
nodes[0].join(None)
for n in nodes[1:]:
    n.join(nodes[0])

# Store some key-value pairs
keys = ["fileA","fileB","fileC"]
for k in keys:
    h = hash_key(k, m)
    nodes[random.randint(0,len(nodes)-1)].store(h, f"Data for {k}")

# Lookup keys
for k in keys:
    h = hash_key(k, m)
    nodes[random.randint(0,len(nodes)-1)].lookup(h)
````

# 2.4.2 Unstructured Peer-to-Peer Systems

**Overview:**  
Unstructured P2P systems feature nodes with ad hoc neighbor lists, forming an overlay network resembling a random graph. Nodes periodically update their neighbors to replace unresponsive nodes. Unlike structured P2P, there is **no deterministic routing**, so data lookup relies on search strategies.

**Data Lookup Methods:**  

1. **Flooding:**  
   - A node sends a request to all neighbors.  
   - Receiving nodes forward the request to their neighbors unless already seen.  
   - If a node has the data, it responds directly or via the forwarding path.  
   - Uses a **Time-To-Live (TTL)** to limit network traffic.  
   - Variants include incrementally increasing TTL to widen search.

2. **Random Walks:**  
   - A node sends a request to a randomly selected neighbor.  
   - Each node forwards the request to another random neighbor until the data is found.  
   - Lower network traffic compared to flooding.  
   - May take longer; multiple simultaneous walks can reduce search time.

3. **Policy-Based Searches:**  
   - Hybrid between flooding and random walks.  
   - Tracks successful responders as preferred neighbors.  
   - Limits flooding to high-degree neighbors for efficiency.

**Efficiency Comparison:**  
- Consider **N** nodes, each data item replicated on **r** nodes:  
  - Average random walk search size: `S ≈ N/r`  
  - Example: If `r/N = 0.1%`, then `S ≈ 1000` nodes.  
  - Flooding (d=10 neighbors, 4 steps) reaches `R(4) ≈ 7290` nodes—much higher than needed at low replication.

**Key Takeaways:**  
- Unstructured P2P systems are flexible but rely on searching rather than deterministic lookups.  
- Flooding ensures completeness but is costly; random walks reduce traffic at the expense of speed.  
- Replication improves search efficiency.  
- Policy-based searches balance efficiency and completeness.


````python
import random
from collections import deque

# ----------------------------
# Node in unstructured P2P
# ----------------------------
class UnstructuredNode:
    def __init__(self, node_id):
        self.node_id = node_id
        self.neighbors = set()
        self.data_store = set()  # data items held

    def add_neighbor(self, other):
        if other != self:
            self.neighbors.add(other)
            other.neighbors.add(self)

    def store_data(self, data_item):
        self.data_store.add(data_item)

    # Flooding search
    def search_flooding(self, data_item, ttl=3, visited=None):
        if visited is None:
            visited = set()
        visited.add(self)
        if data_item in self.data_store:
            return self
        if ttl <= 0:
            return None
        for neighbor in self.neighbors:
            if neighbor not in visited:
                result = neighbor.search_flooding(data_item, ttl-1, visited)
                if result:
                    return result
        return None

    # Random walk search
    def search_random_walk(self, data_item, ttl=5):
        current = self
        for _ in range(ttl):
            if data_item in current.data_store:
                return current
            if not current.neighbors:
                return None
            current = random.choice(list(current.neighbors))
        return None

# ----------------------------
# Build unstructured network
# ----------------------------
N = 20  # number of nodes
nodes = [UnstructuredNode(i) for i in range(N)]

# Randomly connect nodes
for node in nodes:
    neighbors = random.sample(nodes, k=random.randint(2,4))
    for n in neighbors:
        node.add_neighbor(n)

# Replicate some data items
data_items = ["fileA", "fileB", "fileC"]
for item in data_items:
    replicas = random.sample(nodes, k=3)  # each item stored on 3 nodes
    for node in replicas:
        node.store_data(item)

# ----------------------------
# Perform searches
# ----------------------------
# Pick a random node to initiate search
start_node = random.choice(nodes)
target_item = "fileB"

# Flooding search
found_node = start_node.search_flooding(target_item, ttl=4)
if found_node:
    print(f"[Flooding] '{target_item}' found at Node {found_node.node_id}")
else:
    print(f"[Flooding] '{target_item}' not found within TTL")

# Random walk search
found_node_rw = start_node.search_random_walk(target_item, ttl=6)
if found_node_rw:
    print(f"[Random Walk] '{target_item}' found at Node {found_node_rw.node_id}")
else:
    print(f"[Random Walk] '{target_item}' not found within TTL")
````

# 2.4.3 Hierarchically Organized Peer-to-Peer Networks

**Motivation:**  
Unstructured P2P networks face scalability issues due to reliance on flooding or random walks. As networks grow, searches become inefficient. Certain applications, like collaborative CDNs, benefit from **breaking full symmetry** to efficiently locate nodes with resources.

**Super-Peer Concept:**  
- **Super peers:** Special nodes that maintain indexes or act as brokers.  
- **Weak peers:** Regular nodes that rely on super peers for queries and coordination.  
- Forms a hierarchical network:  
  - Weak peers connect to one or more super peers.  
  - Super peers may also form a P2P overlay among themselves.

**Associations Between Weak and Super Peers:**  
- **Fixed association:** Weak peer attaches to a super peer upon joining and remains until leaving.  
- **Backup/redundancy:** Weak peers may connect to multiple super peers for fault tolerance.  
- **Dynamic association:** Weak peers may switch to super peers with more relevant indexes or better lookup performance.

**Challenges:**  
- Choosing which nodes become super peers (leader-election problem).  
- Hierarchical networks trade full symmetry for efficiency but require additional management.

**Key Takeaways:**  
- Hierarchical P2P networks combine the flexibility of unstructured networks with better scalability.  
- Super peers provide efficient indexing and coordination.  
- Dynamic association schemes can further optimize performance.


````python
import random
from sklearn.datasets import load_iris

# ----------------------------
# Load dataset
# ----------------------------
iris = load_iris(as_frame=True)
data = iris.frame
data_items = data.index.tolist()  # use row indices as "data items"

# ----------------------------
# Parameters
# ----------------------------
num_super_peers = 3
num_weak_peers = 10

# ----------------------------
# Super peers
# ----------------------------
super_peers = {i: {"weak_peers": set(), "index": {}, "neighbors": set()} for i in range(num_super_peers)}
for sp_id in super_peers:
    next_sp = (sp_id + 1) % num_super_peers
    super_peers[sp_id]["neighbors"].add(next_sp)
    super_peers[next_sp]["neighbors"].add(sp_id)

# ----------------------------
# Weak peers
# ----------------------------
weak_peers = {}
for wp_id in range(num_super_peers, num_super_peers + num_weak_peers):
    weak_peers[wp_id] = {"super_peer": None, "data": set()}
    sp_id = random.choice(list(super_peers.keys()))
    weak_peers[wp_id]["super_peer"] = sp_id
    super_peers[sp_id]["weak_peers"].add(wp_id)
    # Assign 5 random data items to each weak peer
    assigned_items = random.sample(data_items, k=5)
    for item in assigned_items:
        weak_peers[wp_id]["data"].add(item)
        super_peers[sp_id]["index"][item] = wp_id

# ----------------------------
# Query function
# ----------------------------
def query(weak_peer_id, data_item, visited=None):
    if visited is None:
        visited = set()
    sp_id = weak_peers[weak_peer_id]["super_peer"]
    if sp_id in visited:
        return None
    visited.add(sp_id)
    if data_item in super_peers[sp_id]["index"]:
        return super_peers[sp_id]["index"][data_item]
    for neighbor_id in super_peers[sp_id]["neighbors"]:
        result = query_from_super(neighbor_id, data_item, visited)
        if result:
            return result
    return None

def query_from_super(sp_id, data_item, visited):
    if sp_id in visited:
        return None
    visited.add(sp_id)
    if data_item in super_peers[sp_id]["index"]:
        return super_peers[sp_id]["index"][data_item]
    for neighbor_id in super_peers[sp_id]["neighbors"]:
        result = query_from_super(neighbor_id, data_item, visited)
        if result:
            return result
    return None

# ----------------------------
# Perform a query
# ----------------------------
query_wp = random.choice(list(weak_peers.keys()))
target_item = random.choice(data_items)
result_wp = query(query_wp, target_item)

if result_wp:
    print(f"[Hierarchical P2P with Iris] Weak Peer {query_wp} found row '{target_item}' at Weak Peer {result_wp}")
else:
    print(f"[Hierarchical P2P with Iris] Weak Peer {query_wp} could not find row '{target_item}'")
````

# 2.4.4 Example: BitTorrent

**Overview:**  
BitTorrent is a file-sharing system exemplifying an unstructured peer-to-peer network. Its core principle is to **download files in chunks from multiple peers** and assemble them to obtain the complete file.

**Key Design Goals:**  
- **Collaboration enforcement:** Prevent free riding. Users must upload chunks to others to continue downloading.

**How BitTorrent Works:**  
- **Global Directory / Torrent Files:**  
  - Users access a global directory (e.g., web page) listing torrent files.  
  - Each torrent file contains a link to a tracker and metadata about the file and its chunks.  

- **Tracker Role:**  
  - Tracks active nodes holding chunks of the file.  
  - Typically one tracker per file, but multiple trackers may exist overall.  

- **Downloading and Sharing:**  
  - Nodes begin downloading chunks from peers.  
  - Nodes must share chunks they have with others.  
  - Upload rates are regulated based on reciprocity: peers may slow sending to nodes that download more than they upload.  

- **Peer Selection and Redundancy:**  
  - Nodes receive references to multiple peers, improving download efficiency and redundancy.  

**Centralized vs. Decentralized Aspects:**  
- **Centralized:** Trackers initially coordinate peers for file downloads.  
- **Decentralized / DHT-based:**  
  - Nodes can use structured P2P systems (DHT) to locate peers.  
  - Tracker load is distributed; peers communicate primarily among themselves.  
  - Torrent discovery may use magnet links pointing to DHT lookups.  

**Key Observations:**  
- BitTorrent combines **centralized coordination** with **peer-to-peer decentralization**.  
- Enforces sharing while maintaining efficiency and scalability.  
- Modern implementations minimize reliance on trackers once peers are connected.


````python
import random
from collections import defaultdict

# ----------------------------
# Parameters
# ----------------------------
num_peers = 5
file_name = "iris_dataset"  # represent a dataset as file
num_chunks = 10  # split file into 10 chunks

# ----------------------------
# Tracker (centralized)
# ----------------------------
tracker = defaultdict(set)  # chunk_id -> set of peers having it

# ----------------------------
# Peers
# ----------------------------
peers = {}
for pid in range(num_peers):
    peers[pid] = {
        "chunks": set(),
        "upload": 0,
        "download": 0
    }

# Randomly assign initial chunks to peers (seeding)
for chunk_id in range(num_chunks):
    seed_peer = random.randint(0, num_peers-1)
    peers[seed_peer]["chunks"].add(chunk_id)
    tracker[chunk_id].add(seed_peer)

# ----------------------------
# Peer-to-peer download function
# ----------------------------
def download_chunk(peer_id, chunk_id):
    available_peers = tracker[chunk_id] - {peer_id}
    if not available_peers:
        return False  # no source available
    source_peer = random.choice(list(available_peers))
    # enforce reciprocity: only allow download if source_peer is uploading
    if peers[source_peer]["upload"] >= peers[source_peer]["download"]:
        peers[peer_id]["chunks"].add(chunk_id)
        peers[peer_id]["download"] += 1
        peers[source_peer]["upload"] += 1
        tracker[chunk_id].add(peer_id)
        return True
    return False

# ----------------------------
# Simulate BitTorrent download
# ----------------------------
for peer_id in range(num_peers):
    missing_chunks = set(range(num_chunks)) - peers[peer_id]["chunks"]
    attempts = 0
    while missing_chunks and attempts < 50:
        chunk_to_download = random.choice(list(missing_chunks))
        success = download_chunk(peer_id, chunk_to_download)
        if success:
            missing_chunks.remove(chunk_to_download)
        attempts += 1

# ----------------------------
# Display results
# ----------------------------
for pid, info in peers.items():
    print(f"Peer {pid}: Chunks={sorted(info['chunks'])}, Upload={info['upload']}, Download={info['download']}")

# Check full file completion
for pid, info in peers.items():
    if len(info['chunks']) == num_chunks:
        print(f"Peer {pid} successfully downloaded full file '{file_name}'")
    else:
        print(f"Peer {pid} incomplete: missing {num_chunks - len(info['chunks'])} chunks")
````

# 2.5 Hybrid System Architectures

Real-world distributed systems often **combine multiple architectural styles**, integrating:

- Centralized features  
- Peer-to-peer features  
- Hierarchical organizations  

These hybrid architectures can cross organizational boundaries, producing decentralized and complex solutions.

## 2.5.1 Cloud Computing

Cloud computing evolved from **utility computing**:  
- Customers upload tasks to a data center.  
- Charged per resource usage.  
- Provides **scalable, virtualized resources on demand**.

### Characteristics (Vaquero et al., 2008)
- Easily usable and accessible pool of resources.  
- Dynamic resource allocation based on demand.  
- Pay-per-use model with **Service Level Agreements (SLAs)**.

### Cloud Layers
1. **Hardware Layer**  
   - Physical resources: CPUs, memory, routers, power, cooling.  
   - Typically hidden from end users.  

2. **Infrastructure Layer**  
   - Provides virtual storage and computing resources.  
   - Implemented via virtualization (VMs, virtual disks, etc.).  
   - Backbone for most cloud services.  

3. **Platform Layer**  
   - Offers APIs for developing and deploying cloud applications.  
   - Comparable to an OS for cloud apps.  
   - Examples: Amazon S3 API for storage, computation frameworks.  

4. **Application Layer**  
   - Cloud-hosted applications delivered to end users.  
   - Examples: Gmail, Google Docs, Office 365.  
   - Applications run entirely in the cloud.

### Cloud Service Models

| Service Model | Layer Covered | Examples |
|---------------|---------------|---------|
| IaaS (Infrastructure-as-a-Service) | Hardware + Infrastructure | Amazon EC2, Azure VMs |
| PaaS (Platform-as-a-Service) | Platform | Google App Engine |
| SaaS (Software-as-a-Service) | Application | Gmail, Google Docs, Office 365 |

### Architectural Notes
- Cloud computing is a **highly advanced client-server system**.  
- Server implementation is **hidden from clients**.  
- Servers may be fully distributed across multiple machines.  
- **Function-as-a-Service (FaaS):** Clients execute code without managing servers.


````python
import random
import time
from collections import deque

# ----------------------------
# Parameters
# ----------------------------
num_clients = 5
num_vms = 3  # Infrastructure Layer: virtual machines
tasks_per_client = 3

# ----------------------------
# Infrastructure Layer: VMs
# ----------------------------
vms = [{"id": i, "busy": False, "current_task": None} for i in range(num_vms)]

# ----------------------------
# Platform Layer: Task Queue
# ----------------------------
task_queue = deque()  # tasks waiting for execution
slas = {}  # Track Service Level Agreements (SLA) per task

# ----------------------------
# Clients submit tasks
# ----------------------------
task_id = 0
for client_id in range(num_clients):
    for _ in range(tasks_per_client):
        task = {"task_id": task_id, "client_id": client_id, "duration": random.randint(1,3)}
        task_queue.append(task)
        slas[task_id] = {"start_time": None, "end_time": None, "status": "queued"}
        task_id += 1

# ----------------------------
# Scheduler: Assign tasks to VMs dynamically
# ----------------------------
while task_queue:
    for vm in vms:
        if not vm["busy"] and task_queue:
            task = task_queue.popleft()
            vm["busy"] = True
            vm["current_task"] = task
            slas[task["task_id"]]["start_time"] = time.time()
            # Simulate execution (non-blocking for demo purposes)
            exec_time = task["duration"]
            slas[task["task_id"]]["end_time"] = slas[task["task_id"]]["start_time"] + exec_time
            slas[task["task_id"]]["status"] = "completed"
            print(f"[Cloud] VM {vm['id']} executed Task {task['task_id']} for Client {task['client_id']} (Duration={exec_time}s)")
            vm["busy"] = False
            vm["current_task"] = None

# ----------------------------
# Summary: SLA report
# ----------------------------
print("\n[SLA Report]")
for tid, info in slas.items():
    print(f"Task {tid}: Status={info['status']}, Start={info['start_time']:.2f}, End={info['end_time']:.2f}")
````

# 2.5.2 Edge-Cloud Architecture

Edge computing emerged in response to:

- Increasingly network-connected devices  
- Rise of the Internet-of-Things (IoT)  
- Limitations of purely cloud-based solutions  

**Edge computing** places services “at the edge” of the network, close to devices, complementing cloud computing.

## Conceptual Overview

- Edge lies at the boundary between enterprise networks and the Internet.  
- Examples:  
  - University campus networks with local servers (storage, computing, lectures)  
  - Local services that rarely leave the campus network  
- Edge infrastructures may connect to other local infrastructures or cloud services.  
- The continuum between end devices, edge infrastructures, and cloud infrastructures is often blurred.


## Motivations for Edge Computing

1. **Latency and Bandwidth**  
   - Edge is closer to end devices, reducing latency.  
   - Critical for real-time applications:  
     - Video streaming (reduces jitter)  
     - (Semi-)autonomous driving (local sensor data processing)  
   - Bandwidth issues are less critical today but still relevant for high-volume applications.

2. **Reliability**  
   - Cloud connectivity may be insufficient for mission-critical applications (hospitals, factories, autonomous vehicles).  
   - Edge provides local continuity if cloud connections fail.

3. **Security and Privacy**  
   - Edge can improve security if under a trusted organization’s control.  
   - Regulatory constraints may require data to stay on-premises (e.g., medical records).

## Challenges of Edge Computing

- **Resource limitations:** Less computing, storage, and networking capacity than cloud data centers.  
- **Hardware heterogeneity:** More varied devices than in cloud environments.  
- **Dynamic workloads:** Mobile users, IoT devices, and fluctuating demand.  
- **Orchestration complexity:**  
  - Resource allocation: Guarantee CPU, memory, storage, and network availability.  
  - Service placement: Decide where services run (local edge vs cloud).  
  - Edge selection: Choose the most suitable edge infrastructure for service delivery.  

Managing edge-cloud orchestration is more complex than purely cloud-based systems.

-----

- Edge computing complements the cloud by providing **low-latency, localized services**.  
- Introduces significant challenges in **resource management, service placement, and orchestration**.  
- The edge-cloud continuum is evolving, requiring architectures that balance **cloud and edge capabilities**.


````python
import random
from collections import deque

# ----------------------------
# Parameters
# ----------------------------
num_edge_nodes = 3
num_cloud_nodes = 2
num_clients = 5
tasks_per_client = 3
max_simulation_steps = 20  # prevents infinite loops

# ----------------------------
# Edge and Cloud Nodes
# ----------------------------
edge_nodes = [{"id": i, "capacity": 5, "tasks": [], "latency": 10} for i in range(num_edge_nodes)]
cloud_nodes = [{"id": i, "capacity": 20, "tasks": [], "latency": 50} for i in range(num_cloud_nodes)]
all_nodes = edge_nodes + cloud_nodes

# ----------------------------
# Clients generate tasks
# ----------------------------
task_queue = deque()
task_id = 0
for client_id in range(num_clients):
    for _ in range(tasks_per_client):
        task = {
            "task_id": task_id,
            "client_id": client_id,
            "cpu_demand": random.randint(1,4),
            "latency_sensitive": random.choice([True, False])
        }
        task_queue.append(task)
        task_id += 1

# ----------------------------
# Scheduler: assign tasks with dynamic resource release
# ----------------------------
simulation_step = 0
while task_queue and simulation_step < max_simulation_steps:
    simulation_step += 1
    print(f"\n[Simulation Step {simulation_step}] Tasks in queue: {len(task_queue)}")

    for node in all_nodes:
        # Simulate resource freeing: complete all tasks from previous steps
        node["capacity"] += sum(t["cpu_demand"] for t in node["tasks"])
        node["tasks"] = []

    current_queue_size = len(task_queue)
    for _ in range(current_queue_size):
        task = task_queue.popleft()
        candidate_nodes = edge_nodes if task["latency_sensitive"] else all_nodes
        available_nodes = [n for n in candidate_nodes if n["capacity"] >= task["cpu_demand"]]

        if available_nodes:
            node = random.choice(available_nodes)
            node["tasks"].append(task)
            node["capacity"] -= task["cpu_demand"]
            location = "Edge" if node in edge_nodes else "Cloud"
            print(f"[Scheduler] Assigned Task {task['task_id']} from Client {task['client_id']} to {location} Node {node['id']}")
        else:
            print(f"[Scheduler] Task {task['task_id']} from Client {task['client_id']} delayed (no resources)")
            task_queue.append(task)

# ----------------------------
# Final Summary: Node utilization
# ----------------------------
print("\n[Final Node Utilization]")
for node in all_nodes:
    location = "Edge" if node in edge_nodes else "Cloud"
    print(f"{location} Node {node['id']}: Tasks this step={len(node['tasks'])}, Remaining Capacity={node['capacity']}")

if task_queue:
    print(f"\n[Simulation End] {len(task_queue)} tasks could not be scheduled due to insufficient resources.")
else:
    print("\n[Simulation End] All tasks successfully scheduled.")
````

# 2.5.3 Blockchain Architectures

Blockchain systems are distributed ledgers designed to record transactions without a trusted third party. Applications extend beyond finance to:

- Identity management  
- Resource allocation  
- Electronic voting  
- Health records  

## Key Concepts

### Transactions and Validation
- Transactions (e.g., Alice sends $10 to Bob) must be validated, executed, and stored for auditing.  
- Unlike traditional systems, participants are not inherently trusted; no central authority exists.  

### Blocks and Chain
- Transactions are grouped into **blocks** for efficiency.  
- Each block is **immutable** and appended to a single chain of validated blocks.  
- Blocks can be **replicated** across nodes to simplify verification.

### Consensus
- Determining which node appends the next block requires **distributed consensus**.  
- Approaches:  
  1. **Centralized:** Trusted third party validates transactions.  
  2. **Distributed (permissioned):** Small group of preselected nodes validate transactions; assumes at most $$k \le \frac{n-1}{3}$$ malicious nodes (Byzantine fault tolerance); practical limit: tens of nodes.  
  3. **Decentralized (permissionless):** Any willing node can validate; leader election decides who appends the next block; resource-intensive, so only a fraction of nodes validate at any time.

## Architectural Implications

### Permissioned Blockchains
- Limited number of validators (few tens).  
- Fault-tolerant distributed group, not fully centralized.  
- Trust assumption: only a minority may behave maliciously.

### Permissionless Blockchains
- Fully decentralized in principle.  
- Requires leader election and fairness mechanisms.  
- In practice, validation often concentrates in a few nodes, creating quasi-centralization.

(a) Centralized, (b) Permissioned distributed, (c) Decentralized permissionless blockchain architectures.

----
- Blockchains remove the need for a trusted third party.  
- Core design principles: **immutability, consensus, replication**.  
- Validation mechanisms classify blockchains into centralized, permissioned, and permissionless systems.  
- Even decentralized systems often exhibit **partial centralization** for efficiency.


````python
import hashlib
import random
from collections import deque

# ----------------------------
# Parameters
# ----------------------------
num_nodes = 5
num_transactions = 10
block_capacity = 3  # max transactions per block

# ----------------------------
# Blockchain Node Structure
# ----------------------------
nodes = [{"id": i, "blockchain": [], "pending": deque()} for i in range(num_nodes)]

# ----------------------------
# Generate random transactions
# ----------------------------
for tx_id in range(num_transactions):
    tx = f"Tx{tx_id}: User{random.randint(1,5)} pays User{random.randint(1,5)} ${random.randint(1,100)}"
    # broadcast to all nodes
    for node in nodes:
        node["pending"].append(tx)

# ----------------------------
# Simple consensus: leader election
# ----------------------------
def compute_hash(prev_hash, transactions):
    tx_str = "".join(transactions)
    return hashlib.sha256((prev_hash + tx_str).encode()).hexdigest()

def mine_block(node, prev_hash):
    block_tx = []
    while node["pending"] and len(block_tx) < block_capacity:
        block_tx.append(node["pending"].popleft())
    block_hash = compute_hash(prev_hash, block_tx)
    return {"transactions": block_tx, "prev_hash": prev_hash, "hash": block_hash}

# ----------------------------
# Blockchain simulation
# ----------------------------
# Initial genesis block
genesis_hash = "0"*64
for step in range((num_transactions + block_capacity - 1) // block_capacity):
    # Leader election: choose random node
    leader = random.choice(nodes)
    prev_hash = leader["blockchain"][-1]["hash"] if leader["blockchain"] else genesis_hash
    block = mine_block(leader, prev_hash)
    # Replicate block to all nodes
    for node in nodes:
        node["blockchain"].append(block)
    print(f"[Step {step}] Node {leader['id']} mined block with {len(block['transactions'])} transactions.")

# ----------------------------
# Display final blockchain
# ----------------------------
print("\n[Final Blockchain on Node 0]")
for i, block in enumerate(nodes[0]["blockchain"]):
    print(f"Block {i}: Hash={block['hash'][:10]}..., Transactions={len(block['transactions'])}")
````
